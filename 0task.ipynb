````python
import torch
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv

from numba import njit
from numpy.linalg import norm

from tqdm import tqdm

import seaborn as sns
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
````

````python
df = pd.read_csv(
    "data.ssv", sep=" "
)
````

````python
dts = [0.0001, 0.00025, 0.0005, 0.00075, 0.001, 0.002, 0.004, 0.005, 0.007, 0.01]

len(dts)
````

````python
std_E = []
````

````python
dfs = []
for i in range(len(dts)):
    std_E.append(
        df.iloc[i * 10001: (i + 1) * 10001].TE.std()
    )
    dfs.append(df.iloc[i * 10001: (i + 1) * 10001].TE)    
````

````python
dts = dts[2:]
std_E = std_E[2:]
````

````python
plt.figure(
    figsize=(22, 12)
)

plt.subplot(1, 2, 1)
plt.scatter(
    (np.array(dts)), (np.array(std_E))
)

plt.subplot(1, 2, 2)
plt.scatter(
    np.log(np.array(dts)), np.log(np.array(std_E))
)
plt.xlabel(r'$log(dt)$', fontsize=16)
plt.ylabel(r'$log(\sigma_E)$', fontsize=16)
````
